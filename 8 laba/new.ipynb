```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv('songs.csv')
df
```

```python
df["genre_top"].value_counts()
```

```python
df = df.drop([
    "track_id", 
    "comments", 
    "composer", 
    "favorites", 
    "information", 
    "lyricist", 
    "number", 
    "publisher", 
    "tags"
], axis=1)
df
```

```python
cat_features_mask = (df.dtypes == "object").values
```

```python
num_columns = df.columns[~cat_features_mask]
scaler = StandardScaler()
for col in num_columns:
    df[col] = scaler.fit_transform(df[col].values.reshape(-1, 1))
```

```python
cat_columns = df.columns[cat_features_mask]
label_encoder = LabelEncoder()
for col in cat_columns:
    df[col] = label_encoder.fit_transform(df[col])
```

```python
df
```

```python
x = df.drop(["genre_top"], axis=1)
y = df["genre_top"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
dtc = DecisionTreeClassifier(random_state=10)
dtc.fit(x_train, y_train)

pred = dtc.predict(x_test)
```

```python
class_rep_tree = classification_report(y_test, pred)

print("-------------------------------------------------------")
print("Decision Tree: \n", class_rep_tree)
print("-------------------------------------------------------")
```
